# 1. Data Cleansing
- Memuat file "review-lazada.csv
- Menghapus data dengan nilai **null** dari kolom reviewTitle, reviewContent, dan boughtDate
- Memastikan tidak ada data yang double
- Mengekspor data yang sudah dibersihkan ke dalam file CSV baru

```python
df = pd.read_csv('/content/review-lazada.csv')

df_cleaned = df.dropna(subset=['reviewTitle', 'reviewContent', 'boughtDate'])

df = df.drop_duplicates()

output_cleaned_path = 'cleaned-review-lazada.csv'
df_cleaned.to_csv(output_cleaned_path, index=False)

df_cleaned.head()
```

# 2. Sentiment Analysis
- Memuat file yang sudah dibersihkan oleh flow data cleansing
- Membuat dan menentukan sentimen berdasarkan kolom rating
- Mengekspor hasil analisis sentimen ke file csv baru

```python
df_cleaned = pd.read_csv(output_cleaned_path)

def sentiment(row):
    if row['rating'] >= 4:
        return 'Positive'
    elif row['rating'] == 3:
        return 'Neutral'
    else:
        return 'Negative'

df_cleaned['sentiment'] = df_cleaned.apply(sentiment, axis=1)

output_sentiment_path = 'sentiment-review-lazada.csv'
df_cleaned.to_csv(output_sentiment_path, index=False)

df_cleaned.head()
```
